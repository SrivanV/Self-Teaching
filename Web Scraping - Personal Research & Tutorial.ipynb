```python
import bs4
import lxml
import requests
```

```python
# CONSTRUCT A BASIC REQUEST
Res = requests.get('https://www.youtube.com/')
```

```python
# OUTPUT THE REQUEST RESULT
Res.text
```

```python
# USE BeautifulSoup TO CREATE AN OBJECT THAT CAN PARSE THROUGH THE REQUEST RESULTS FROM EARLIER
Soup = bs4.BeautifulSoup(Res.text, 'lxml')
Soup
```

```python
# A UNIQUE DATA TYPE FOR THE OBJECT. GOOD TO KEEP IN MIND AS I MOVE FORWARD!
type(Soup)
```

```python
# TEST TO SEE THAT IT'S WORKING
TitleTag = Soup.select('Title')
TitleTag
```

```python
# INSTINCT TOLD ME THAT IT SHOULD JUST BE A STRING BUT THAT'S NOT THE CASE
type(TitleTag)
```

```python
# ANOTHER LAYER INTO THE TEST. A RE-AFFIRMATION OF THE FACT THAT IT'S NOT JUST A STRING
TitleTag[0]
```

```python
# FINAL STEP TO RETREIVE THE ACTUAL STRING THAT I WANTED
TitleTag[0].getText()
```

```python
# ANOTHER REQUEST TO EXTRACT TEXT FROM A VERY COMMONLY USED WEBSITE LIKE Wikipedia
Res2 = requests.get('https://en.wikipedia.org/wiki/Room_641A')
```

```python
# USE THE SAME BASIC CONCEPT FROM EARLIER
Soup2 = bs4.BeautifulSoup(Res2.text, 'lxml')
```

```python
# EXTRACT PARTICULAR PARTS OF THE WEBPAGE TO HAVE AN UNDERSTANDING OF HOW TO NAVIGATE THE REQUEST RESULT
Soup2.select('.mw-headline')
```

```python
# MAKE THE RESULTS MORE VISUALLY READABLE
for Titles in Soup2.select('.mw-headline'):
    print(Titles.text)
```

```python
# A SELF-ASSIGNED TEST TO MAKE SURE THAT I FULLY UNDERSTAND THE CONCEPT.
# MY GOAL IS TO RETREIVE A POPULAR IMAGE ASSOCIATED WITH A VERY MYSTERIOUS CRYPTOGRAPHY CHALLENGE POSTED ONLINE A FEW YEARS AGO
Cicada3301 = requests.get('https://en.wikipedia.org/wiki/Cicada_3301')
```

```python
# bs4 OBJECT FOR PARSING
CSoup = bs4.BeautifulSoup(Cicada3301.text, 'lxml')
```

```python
# ISOLATE THE IMAGE SECTION
Logo = CSoup.select('.thumbimage')
Logo
```

```python
# CHECK THE CLASS TO DOUBLE CHECK THAT I GRABBED THE RIGHT PORTION
Logo[0]['class']
```

```python
# GET THE DIRECT LINK OF THE IMAGE
Logo[0]['src']
```

```python
# PREPENDING THE PROTOCOL IN ORDER TO CREATE A FULL URL
LogoLink = 'http:' + Logo[0]['src']
LogoLink
```

```python
# USE THE NOW COMPLETE URL IN ORDER TO DIRECTLY REQUEST THE END GOAL THAT I'M AFTER
Cicada = requests.get(LogoLink, 'lxml')
Cicada
```

```python
# EXPLORE EXACTLY WHAT WE CALLED BACK FROM THE STEPS I'VE TAKEN SO FAR
# AFTER A BIT OF RESEARCH, I DISCOVERED THAT THE CONENTS ARE INDIVIDUAL PIXEL TAGS THAT NEED TO BE PROCESSED FURTHER
Cicada.content
```

```python
# AGAIN I CONFIRM THAT THE RESULT ABOVE IS NOT THE IMAGE DIRECTLY 
type(Cicada)
```

```python
# CREATING AN IMAGE THAT I CAN WRITE THE CONTENT OF THE Cicada.content RESULTS  INTO
CicadaLogo = open('Cicada.jpg', 'wb')
```

```python
#CALL THE .write FUNCTION IN ORDER TO FINALLY GET MY DESIRED IMAGE 
CicadaLogo.write(Cicada.content)
```

```python
# BE SURE TO CLOSE WHEN DONE WRITING
CicadaLogo.close()
```
